```python

 
import re

def getHash(str):
   h=0
   for c in str:
      # print(c,ord(c))
      h+=ord(c)
      h*=17
      # print("*17=", h)
      h=h%256
      # print("%256=", h)
   # print("Got ",h)
   return h

def addOrUpdate(boxes, command):
    lens = command[-1]
    label = command[:-2]
    box = getHash(label)
    for l in boxes[box]:
        if l[0] == label:
            l[1] = int(lens)
            return
    boxes[box].append([label, int(lens)])

def remove(boxes, command):
    label = command[:-1]
    box = getHash(label)
    r = [i for i, b in enumerate(boxes[box]) if b[0] == label]
    if len(r) == 1:
       boxes[box].pop(r[0])
    return


def answer(input):
    total=0
    boxes = [[] for _ in range(256)]
    for s in input.split(","):
    #   print(s)
      if "=" in s:
        addOrUpdate(boxes, s)
      elif "-" in s:
        remove(boxes, s)

    for boxNum, box in enumerate(boxes):
       if(len(box) >0 ):
            # print(f"Box {boxNum}")
            # print(box)
            focals = [(boxNum+1)*(slotNum+1)*slot[1] for slotNum, slot in enumerate(box)]
            # print(focals)
            total+=sum(focals)


    return total

   

with open("input.txt") as f:
   print(answer(f.read()))

    
```
