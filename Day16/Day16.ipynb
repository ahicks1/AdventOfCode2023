```python

 
import re

def traceBeam(maze, location, delta):
   x,y = location
   dX, dY = delta
   pipe = maze[y][x]
   if pipe == "/":
      dX,dY = -dY,-dX
   elif pipe == "\\":
      dX,dY = dY,dX
   elif (pipe == "-" and dY != 0) or(pipe == "|" and dX != 0) :
      p1 = ((x-dY,y-dX),(-dY,-dX))
      p2 = ((x+dY,y+dX),(dY,dX))
      return [p1,p2]
   return [((x+dX,y+dY), (dX,dY))]

def inBounds(maze, position):
   x,y = position
   xSize = len(maze[0])
   ySize = len(maze)
   return x < xSize and x > -1 and y < ySize and y > -1

def printMaze(maze):
   print("\n".join(["".join(["#" if e else "." for e in line]) for line in maze]))

def getAllStarts(input):
   beams = []
   l = len(input)
   for i in range(len(input)):
      beams.append(((0  ,i),( 1,0)))
      beams.append(((l-1,i),(-1,0)))
      beams.append(((i,  0),(0, 1)))
      beams.append(((i,l-1),(0,-1)))
   return beams

def answer(input):
   solutions = []
   # beams = [((0,0),(1,0))]
   for b in getAllStarts(input):
      beams = [b]
      print(beams)
      energized = [[0]*len(input[0]) for _ in range(len(input))]
      seen = set()
      updated = True
      while updated :
         # print("step")
         # printMaze(energized)
         updated = False
         for position, delta in beams:
            x,y = position
            if (position,delta) not in seen:
               updated = True
               seen.add((position,delta))
               energized[y][x] = 1  
         beams = [(newPos, newDelta) for position, delta in beams for newPos, newDelta in traceBeam(input, position, delta) if inBounds(input, newPos) ]
         # print(beams)
      solutions.append(sum([x for y in energized for x in y]))
      # print("\n".join(energized))
   printMaze(energized)
   return max(solutions)

   

with open("input.txt") as f:
   print(answer(f.read().split("\n")))

    
```
